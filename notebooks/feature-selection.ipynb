```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.decomposition import PCA
```

```python
from ucimlrepo import fetch_ucirepo 
  
dry_bean = fetch_ucirepo(id=53)
  
X = dry_bean.data.features 
y = dry_bean.data.targets 
```

```python
X
```

```python
y
```

```python
sfm = SelectFromModel(RandomForestClassifier()).fit(X, y)
```

```python
X_sfm = pd.DataFrame(sfm.transform(X), columns=sfm.get_feature_names_out())
```

```python
X_sfm
```

```python
pca = PCA().fit(X)
```

```python
list(map(lambda p: round(p,5), pca.explained_variance_ratio_))

# 2 dimmensions -> 97% variance
```

```python
pca = PCA(n_components=2).fit(X)
```

```python
pd.DataFrame(pca.transform(X), columns=pca.get_feature_names_out())
```
